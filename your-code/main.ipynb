#### 2.1. List all language names and number of related articles in the order they appear in wikipedia.org.

````python
# This is the url you will scrape in this exercise
url = 'https://www.wikipedia.org/'
````

````python
response = requests.get(url)
response
````

````python
soup = BeautifulSoup(response.content, "html.parser")
soup
````

````python
languages = soup.find_all('div', attrs={"class": ['central-featured-lang']})
````

````python
for language in languages:
    #print(f"{language.find('strong').text} | {language.find('small').text}")
    # Get the number of articles from the <small> tag
    article_text = language.find('small').text.strip()
    # Extract just the number part
    match = re.search(r'([\d][\d,.\s]*)', article_text)
    num_articles = match.group(1) if match else None

    print(f"Language: {language.find('strong').text}")
    print(f"Number of articles: {num_articles}")
````

#### 3. Display Metacritic top 24 Best TV Shows of all time (TV Show name, initial release date, metascore rating, film rating system and description) as a pandas dataframe.
Hint: If you hover over the title of the movie, you should see the director's name. Can you find where it's stored in the html?

````python
# This is the url you will scrape in this exercise 
url = 'https://www.metacritic.com/browse/tv/'
````

````python
headers = {
       "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36"
   }
````

````python
response = requests.get(url, headers=headers)
response
````

````python
soup = BeautifulSoup(response.content, "html.parser")
````

````python
shows = soup.find_all("div", attrs={"class":"c-finderProductCard"})
````

````python
titles = []
dates = []
scores = []
descriptions = []
show_links = []
img_src = []

for show in shows:
    # titles
    name = show.find(class_='c-finderProductCard_titleHeading').text
    name = re.sub(r'^\d+\.\s*', '', name)
    titles.append(name)

    date = show.find(class_='c-finderProductCard_meta').get_text(strip=True)
    dates.append(date.split("•")[0])

    scores.append(show.find(class_='c-siteReviewScore').get_text(strip=True))

    descriptions.append(show.find(class_='c-finderProductCard_description').get_text(strip=True))

    show_links.append('https://www.metacritic.com/'+show.find('a', class_='c-finderProductCard_container').get('href'))

    img_src.append(show.find('img'))

df = pd.DataFrame({
    "titles": titles,
    "release_date": dates,
    "rating": scores,
    "description":descriptions,
    "show_links":show_links,
    "img_source": img_src
})

df
````
